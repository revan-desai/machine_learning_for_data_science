```python
import seaborn as sns
import pandas as pd

datatip=sns.load_dataset('tips')
sns.scatterplot(x='total_bill',y='tip',data=datatip)   # bivariate analysis
```

```python

sns.scatterplot(x='total_bill',y='tip',hue='sex',style='smoker',size='size',data=datatip) # size is also column name in dataset
 # multivariate analysis
```

```python
tita=sns.load_dataset('titanic')

tita.head(1)
```

```python
sns.barplot(x='pclass',y='age',hue='sex', data=tita)
```

```python
sns.boxplot(x='sex',y='age',hue='survived',data=tita)
```

```python
sns.distplot(tita['age'])
```

```python
sns.kdeplot(tita[tita['survived']==0]['age'])
sns.kdeplot(tita[tita['survived']==1]['age'])
```

```python
pd.crosstab(tita['pclass'],tita['survived'])
```

```python
sns.heatmap(pd.crosstab(tita['pclass'],tita['survived']))
```

```python
tita.groupby('sex').mean()
```

```python
tita.groupby('pclass').mean()
```

```python
tita.groupby('pclass').mean()['survived']*100  # percentage of survived people from each class
```

```python
(tita.groupby('pclass').mean()['survived']*100).plot(kind='pie')
```

```python
datairis=sns.load_dataset('iris')
sns.pairplot(datairis,hue='species')
```

```python
flightdata=sns.load_dataset('flights')

flightdata.groupby('year').sum().head(5)
```

```python
sns.lineplot(x='year',y='passengers',data=flightdata)
```

```python
flightdata.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(flightdata.pivot_table(index='month',columns='year',values='passengers'))

# observe at centre values are high passengers are travellling more during june july august sept
```

```python
sns.clustermap(flightdata.pivot_table(index='month',columns='year',values='passengers'))
```
